```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv( "C:\\Users\\leo\\Desktop\\POWER BI DATA\\ALL DATA\\Data Analysis Projects\\Diwali Sales Data.csv" , encoding= "unicode_escape")
```

```python
df.shape
df.head(10)
df.info()
```

```python
df.info()
df.drop(['Status', 'unnamed1'], axis = 1, inplace=True)

```

```python
df.isnull()
df.isnull().sum
df.dropna(inplace=True)
df.rename(columns={'Marital_Status': 'Shaadi'})
```

```python
df['Amount']  = df['Amount'].astype('int')
df['Amount'].dtypes
df.describe()
```

Gender

```python

ax= sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
sns.barplot(x= 'Gender', y='Amount', data= sales_gen)
```

```python
sns.countplot(data= df, x='Age Group', hue= 'Gender')
```

```python
ax = sns.countplot(data= df, x='Age Group', hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
order_gen = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending=False)
sns.set(rc={'figure.figsize': (15,12)})
sns.lineplot(data = order_gen , x = 'State', y= 'Orders')
```

```python
order_gen = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize': (15,12)})
sns.barplot(data = order_gen , x = 'State', y= 'Orders' , hue= 'Orders')
```

```python
ax = sns.countplot(data = df , x = 'Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by ='Amount',ascending=False)
sns.barplot(data = sales , x = 'MaritaL_Status', y= 'Amount' , hue= 'Gender')
```
